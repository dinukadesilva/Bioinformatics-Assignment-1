**2) Consider random sequences over the alphabet {R, Y}, where in each position R is drawn with probability p, and Y is drawn with probability q = 1-p.**

**(i)   Determine the expected waiting time to pattern RR as a function of the parameter p. [2 pnts]**

[Equation-1] w(R) = 1 + q*w(Y)  

[Equation-2] w(Y) = 1 + p*w(R) + q*w(Y)

By inserting Equation-1 to Equation-2,

w(Y) = 1 + p*(1 + q*w(Y)) + q*w(Y)

w(Y) = 1 + p + p*q*w(Y) + q*w(Y)

w(Y) = (1 + p) / (1 - p*(1-p) - (1-p))

w(Y) = (1 + p) / (1 - p + p^2 - 1 + p)

w(Y) = (1 + p) / p^2

The expected waiting time to pattern RR is (1 + p) / p^2

w(Y) = 1 + (p / (1 - p)) + q*w(Y)

```python
import numpy as np
from random import random

'''
Waiting Time:
The time it would take until a particular pattern is first encountered in a sequence.
By time, we usually mean how long a sequence has to be generated till we see our pattern of interest.
'''
def waiting_time(composition, pattern) :

    # Sanity check on composition
    assert abs( sum(composition.values())-1 ) < 0.01 , 'Probabilities do not add up to 1.'

    # Sanity check on pattern
    for letter in pattern :
        assert letter in composition, 'Pattern involves letters not specified in composition'

    pattern_size = len(pattern)
    sequence =''

    # Main loop
    while len(sequence) < pattern_size or sequence[-pattern_size:] != pattern :
        ran = random()
        p = 0
        for letter in probability :
            p += probability[letter]
            if ran<p:
                sequence += letter
                break

    # Return length of the generated sequence
    return len(sequence)

# Defining the composition
probability = dict()
probability['R']=0.5
probability['Y']=0.5
pattern = "RYY"

def mean_waiting_time(probability, n, pattern):
    return np.mean([waiting_time(probability, pattern) for i in range(n)])

print("[p=0.5]  Mean waiting time for RR of 10 sequences: %.3f" % mean_waiting_time({"R": 0.5, "Y": 0.5}, 10, "RR"))
print("[p=0.75] Mean waiting time for RR of 10 sequences: %.3f" % mean_waiting_time({"R": 0.75, "Y": 0.25}, 10, "RR"))
print("[p=0.5]  Mean waiting time for RY of 10 sequences: %.3f" % mean_waiting_time({"R": 0.5, "Y": 0.5}, 10, "RY"))
print("[p=0.75] Mean waiting time for RY of 10 sequences: %.3f" % mean_waiting_time({"R": 0.75, "Y": 0.25}, 10, "RY"))

print("[p=0.5]  Mean waiting time for RR of 100 sequences: %.3f" % mean_waiting_time({"R": 0.5, "Y": 0.5}, 100, "RR"))
print("[p=0.75] Mean waiting time for RR of 100 sequences: %.3f" % mean_waiting_time({"R": 0.75, "Y": 0.25}, 100, "RR"))
print("[p=0.5]  Mean waiting time for RY of 100 sequences: %.3f" % mean_waiting_time({"R": 0.5, "Y": 0.5}, 100, "RY"))
print("[p=0.75] Mean waiting time for RY of 100 sequences: %.3f" % mean_waiting_time({"R": 0.75, "Y": 0.25}, 100, "RY"))

print("[p=0.5]  Mean waiting time for RR of 1000 sequences: %.3f" % mean_waiting_time({"R": 0.5, "Y": 0.5}, 1000, "RR"))
print("[p=0.75] Mean waiting time for RR of 1000 sequences: %.3f" % mean_waiting_time({"R": 0.75, "Y": 0.25}, 1000, "RR"))
print("[p=0.5]  Mean waiting time for RY of 1000 sequences: %.3f" % mean_waiting_time({"R": 0.5, "Y": 0.5}, 1000, "RY"))
print("[p=0.75] Mean waiting time for RY of 1000 sequences: %.3f" % mean_waiting_time({"R": 0.75, "Y": 0.25}, 1000, "RY"))

print("[p=0.5]  Mean waiting time for RR of 10000 sequences: %.3f" % mean_waiting_time({"R": 0.5, "Y": 0.5}, 10000, "RR"))
print("[p=0.75] Mean waiting time for RR of 10000 sequences: %.3f" % mean_waiting_time({"R": 0.75, "Y": 0.25}, 10000, "RR"))
print("[p=0.5]  Mean waiting time for RY of 10000 sequences: %.3f" % mean_waiting_time({"R": 0.5, "Y": 0.5}, 10000, "RY"))
print("[p=0.75] Mean waiting time for RY of 10000 sequences: %.3f" % mean_waiting_time({"R": 0.75, "Y": 0.25}, 10000, "RY"))
```
